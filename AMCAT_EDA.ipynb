```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming data is your DataFrame
n_cols = len(data.columns)
n_rows = (n_cols + 2) // 3  # Calculate number of rows needed for 3 columns per row

fig, axes = plt.subplots(n_rows, 3, figsize=(20, n_rows * 6))
axes = axes.flatten()  # Flatten the axes array for easier indexing

# Define a function to plot data based on its type
def plot_data(column, ax):
    if pd.api.types.is_numeric_dtype(data[column]):
        sns.histplot(data[column], kde=True, ax=ax)
        ax.set_title(f'Distribution of {column} (Numerical)')
    elif data[column].dtype == 'object' or data[column].dtype.name == 'category':
        sns.countplot(x=column, data=data, ax=ax)
        ax.set_title(f'Distribution of {column} (Categorical)')
    elif pd.api.types.is_datetime64_any_dtype(data[column]):
        data[column] = pd.to_datetime(data[column])  # Ensure it's datetime
        data[column].value_counts().sort_index().plot(ax=ax)
        ax.set_title(f'Time Distribution of {column} (Datetime)')
        ax.set_xlabel('Date')
        ax.set_ylabel('Count')

# Iterate through each column and its corresponding axis
for i, col in enumerate(data.columns):
    plot_data(col, axes[i])  # Call the plotting function

# Hide any unused axes
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

## Times of India article dated Jan 18, 2019 states that â€œAfter doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.â€ 

```python
from scipy import stats

# Define the relevant job roles to analyze
relevant_roles = ['Programmer Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']

# Filter the dataset for the specified job roles
filtered_data = data[data['Designation'].isin(relevant_roles)]

# Extract the salary data for the filtered roles
salary_data = filtered_data['Salary']

# Define the claimed average salary in actual units (e.g., converting lakhs to the full value)
claimed_mean_salary = 2.75 * 100000  # 2.75 lakhs = 275000

# Perform a one-sample t-test to compare the sample mean to the claimed mean
t_stat, p_value = stats.ttest_1samp(salary_data, claimed_mean_salary)

# Print out the results
print(f"Mean Salary of Selected Roles: {salary_data.mean():.2f}")
print(f"Claimed Mean Salary: {claimed_mean_salary:.2f}")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Set the significance level
alpha = 0.05  

# Determine whether to reject or fail to reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis: The average salary significantly differs from the claimed mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the average salary and the claimed mean.")
```


# Conclusion
The analysis of the AMCAT dataset offers valuable insights into salary patterns, specializations, and skill sets of recent graduates in various positions. Below are some significant findings:

## Salary Trends:
Statistical tests indicate that the average salary for roles such as Programming Analyst, Software Engineer, Hardware Engineer, and Associate Engineer aligns with the figures mentioned in the Times of India article. There is no notable difference between the reported salaries and the actual data, suggesting that industry standards are accurately reflected for these positions.

## Influence of Specialization:
Graduates specializing in Computer Science and IT-related fields tend to achieve higher salaries, highlighting the strong demand for these skill sets within the technology sector.

## Gender Representation:
The dataset reveals a disproportionate distribution of male and female graduates across different job roles, indicating possible gender biases or disparities in specific specializations and positions.

## Skill Assessment:
Skills such as programming, computer science, and other technical competencies show a positive correlation with salary, underscoring their significance in securing higher compensation. Additionally, soft skills like conscientiousness, agreeableness, and openness to experience also demonstrate a moderate correlation with job performance and salary, emphasizing the importance of interpersonal skills.

## Educational Background:
Graduates from Tier 1 colleges tend to command higher salaries compared to those from Tier 2 or Tier 3 institutions. This trend underscores the influence of college reputation on initial employment opportunities and salary levels.
